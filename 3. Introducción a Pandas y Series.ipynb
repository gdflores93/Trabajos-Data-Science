### Ejercicio 1: Usar el .CSV para construir un objeto de series de tiempo con el indice igual a los años de cada juego olímpico.

```python
import pandas as pd
df = pd.read_csv('./Olympic.csv',encoding='latin-1')
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df2=df.set_index('Year')
df2
```

### Ejercicio 2: Separar nombres de apellidos en dos columnas distintas usando la coma como separador.

```python
import pandas as pd
df2=df['Athlete'].str.split(',', n=1, expand=True)
df2.columns=['Athlete Surname','Athlete First Name']
df3=pd.concat([df,df2],axis=1)
df3.drop(['Athlete'], axis = 'columns', inplace=True)
df3.reindex(columns=['Year','City','Sport','Discipline','Event','Gender','Athlete Surname','Athlete First Name','Gender','Event_Gender','Country','Country_Code','Medal'])
```
